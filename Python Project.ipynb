```python
def convert_units(value, from_unit, to_unit, conversion_factors):
    if from_unit not in conversion_factors or to_unit not in conversion_factors:
        raise ValueError("Unsupported units.")

    # Convert to base unit, then to the target unit
    return value * conversion_factors[from_unit] / conversion_factors[to_unit]

def main():
    print("Simple Unit Converter")
    print("1. Length Conversion")
    print("2. Weight Conversion")

    try:
        choice = int(input("Choose the type of conversion (1/2): "))

        if choice == 1:
            conversion_factors = {
                "meters": 1,
                "kilometers": 1000,
                "centimeters": 0.01,
                "millimeters": 0.001,
                "miles": 1609.34,
                "yards": 0.9144,
                "feet": 0.3048,
                "inches": 0.0254
            }
        elif choice == 2:
            conversion_factors = {
                "grams": 1,
                "kilograms": 1000,
                "milligrams": 0.001,
                "pounds": 453.592,
                "ounces": 28.3495
            }
        else:
            print("Invalid choice. Please select 1 or 2.")
            return

        value = float(input("Enter the value to convert: "))
        from_unit = input("Enter the from unit: ").lower()
        to_unit = input("Enter the to unit: ").lower()

        result = convert_units(value, from_unit, to_unit, conversion_factors)
        print(f"{value} {from_unit} is equal to {result:.4f} {to_unit}")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
